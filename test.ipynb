```python
# 1. Add Month column as string
df["Month"] = df["Date"].dt.to_period("M").astype(str)

# 2. Group by Month and sum
monthly_expense = df.groupby("Month")["Amount"].sum().reset_index()

# 3. Plot bar chart
import plotly.express as px

fig = px.bar(
    monthly_expense,
    x="Month",
    y="Amount",
    title="Total Expense per Month",
    text="Amount"
)

fig.show()
```
